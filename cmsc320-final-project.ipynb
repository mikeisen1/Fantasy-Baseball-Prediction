<h2>Introduction</h2>

As the National Pastime, baseball has remained a staple of American culture ever since the 1840s. As the years have progressed and technology has advanced, baseball found itself as the early and current center of statistics in sports. This emergence as the leader of the statistical revolution slowly began in 1913, with the establishment of the record keeping company, <a href="http://www.esb.com/about/" target="_blank">Elias Sports Bureau</a>, and in 1947 with the hire of <a href="https://www.baseball-reference.com/bullpen/Allan_Roth" target="_blank">Allan Roth</a> by the then Brooklyn Dodgers (who later moved to their current home in Los Angeles). While the Elias Sports Bureau and Roth were invaluable in promoting the use of statistics in the sport, baseball starting using traditional and advanced statistics in earnest in the 1970s, with the establishment of the Society for American Baseball Research (<a href="https://sabr.org/about/founders" target="_blank">SABR</a>) in 1971 and <a href="https://baseball.fandom.com/wiki/Bill_James" target="_blank">Bill James</a>'s first publication of The Bill James Baseball Abstracts in 1977. The growth of Major League Baseball's use of statistical accelerated even more with the success of the Oakland Athletics in the early 2000s, who used player data to help them adhere to the <a href="https://grantland.com/features/the-economics-moneyball/" target="_blank">Moneyball</a> philosophy.

At a similar time of the exponential increase of statistical use in MLB, fantasy sports, including fantasy baseball, has grown in its popularity. In fantasy baseball, any number of people can form a fantasy league together, with a commissioner to oversee the league operations. Each member of the league, including the commissioner, assumes control of their own team and attempts to select, through a league draft, as many of the best players in baseball as possible. In fantasy baseball, "best" or "good" players can be determined in any number of ways, with the most common being through the points system. In the points system, players can earn a fantasy team points through positive individual actions in a game, such as getting a hit or scoring a run, and can lose points through negative individual actions in a game, such as a batter striking out or a pitcher giving up a hit. Throughout the season, a team must accumulate as many points as possible. This can be achieved by trading players with other league members (team owners) or by adding a player from the league free agent pool and dropping a player currently on the team, giving each owner as close to a feel of running a real life team as possible. At the end of the season, one team and its owner is crowned the fantasy league champion.

Many fantasy baseball leagues are single-season leagues, meaning every year the team owners select new teams from all of the players playing in MLB, with some players from the <a href="https://www.milb.com/about" target="_blank">minors</a>, being available in a draft. However, an increasing trend in fantasy sports is the concept of dynasty leagues, which provide an even closer feel of controlling a real-life team. A dynasty baseball league is long-term, meaning that there is an initial draft at the inception of the league and at the end of a season, each owner selects a certain number of players on their team they want to keep for the following season. There is then a draft of the remaining players in the player pool and teams can even trade current and future draft picks, just like teams do in professional sports. Fantasy baseball is perfect for anyone interested in baseball and statistics, as it relies on statistics to determine the point values of players, which helps determine the strongest fantasy league team at the end of the season. The points settings used to calculate the number of points produced by a player is listed under the section "Points System".

<h2>Data Import and Processing</h2>

First, we must obtain the data, which involves fantasy baseball statistics for both batters and pithcers for 10 years. Since the tutorial is in the context of my dynasty baseball league, We Are Sabermetrics, the statistics and points data must come from the league website. When accessing the player data, there were three categories of statistics that were of interest: Standard, Extra, and Sabermetric. The players were separated into hitters and pitchers due to the different stats applicable to both groups of players. Then, for each year and group of players, a csv file was downloaded from the league website and then read into a Pandas DataFrame.

<img src="import-processing.jpg" alt="2020 Batters Standard">

After uploading the data, we first had to combine the data for each of batters and pitchers as appropriate. For each year, there were three csv files each for batters and pitchers (six in total for a year), with each file providing some different set of player statistics. For ease of future analysis and to avoid storing duplicate data, the dataframes containing batter data must be merged together and the dataframes containing pitcher data must be merged together. At the end of this step, we have five (key, value) pairs in the stats dictionary (which contains the data for both batters and pitchers), where each key is a unique year and each value is a list containing the merged batter data and the merged pitcher data for that year.

TEXT FILES FOR DATA DICTIONARIES!!!!!

```python
#will contain the dataframes of the data from the csv files
stats = dict()

#find the current working directory, used in reading the csv files
cwd = os.getcwd()

#reading the csv files into dataframes and adding them to the stats dictionary
for i in range(2016,2021):
    stats[i] = []
    stats[i].append(pd.read_csv(cwd + "/" + str(i) + "/Fantrax-Players-We Love Sabermetrics.csv"))
    for j in range(1,6):
        file =  cwd + "/" + str(i) + "/Fantrax-Players-We Love Sabermetrics (" + str(j) + ").csv"
        stats[i].append(pd.read_csv(file))

#all years of the data, for similar category (Tracked, Extra, Sabermetric) and player group (batter or pitcher)
#have the exact same columns names. We want to remove duplicate columns and the following dropi variables
#contain the columns to be dropped
drop1 = ((set(stats[2020][0].columns) & set(stats[2020][1].columns)) |
         (set(stats[2020][1].columns) & set(stats[2020][2].columns))).difference(['Player','FPts'])

drop2 = ((set(stats[2020][0].columns) & set(stats[2020][2].columns)) |
         (set(stats[2020][1].columns) & set(stats[2020][2].columns))).difference(['Player','FPts'])
    
drop4 = ((set(stats[2020][3].columns) & set(stats[2020][4].columns)) |
         (set(stats[2020][4].columns) & set(stats[2020][5].columns))).difference(['Player','FPts'])
    
drop5 = ((set(stats[2020][3].columns) & set(stats[2020][5].columns)) |
         (set(stats[2020][4].columns) & set(stats[2020][5].columns))).difference(['Player','FPts'])   

#dropping duplicate columns and merging the batters data together and then the pitchers data together, leaving
#one dataframe for batters and one for pitchers for each year 2011-2020.
def convert_to_float(df,column):
    df[column] = df[column].apply(lambda x : float(x))
    
def convert_to_int(df,column):
    df[column] = df[column].apply(lambda x : int(x))

#dropping the columns that are not needed
for key in stats.keys():
    for i in range(6):
        if i == 1:
            stats[key][i].drop(drop1,axis=1,inplace=True)
        elif i == 2:
            stats[key][i].drop(drop2,axis=1,inplace=True)
        elif i == 4:
            stats[key][i].drop(drop4,axis=1,inplace=True)
        elif i == 5:
            stats[key][i].drop(drop5,axis=1,inplace=True)
        else:
            stats[key][i].drop(["Opponent","Team"],axis=1,inplace=True)
    batters = stats[key][0].merge(stats[key][1],on=['Player','FPts']).merge(stats[key][2],on=['Player','FPts'])
    pitchers = stats[key][3].merge(stats[key][4],on=['Player','FPts']).merge(stats[key][5],on=['Player','FPts'])
    stats[key] = [batters,pitchers]
```

```python
#creating the years column and filtering out batters who do not have enough at-bats
for i in range(2016,2021):
    if i == 2020:
        stats[2020][0] = stats[2020][0][stats[2020][0]['AB'] >= 50]
    else:
        stats[i][0] = stats[i][0][stats[i][0]['AB'] >= 125]
    stats[i][0]['year'] = i
    stats[i][1]['year'] = i
    
#age is initially constant for a player regardless of year, so that is fixed here
for year in range(2016,2020):
    stats[year][0]['Age'] = stats[year][0]['Age'] - (2020 - year)
    stats[year][1]['Age'] = stats[year][1]['Age'] - (2020 - year)

#deep copies and appends each dataframe in a dictionary of dataframes
def get_all_bp(players_data, group, start_year=2016,end_year=2021):
    all_data = copy.deepcopy(players_data[start_year][group])
    for year in range(start_year + 1, end_year):
        all_data = all_data.append(copy.deepcopy(players_data[year][group]),ignore_index=True)
    return all_data

#combining the stats of all batters into 1 dataframe and removing more not needed columns
all_batters = get_all_bp(stats,0)
all_batters.drop(['CYC','Sl','GIDP','SF','SH','IBB','HBP','Status'],axis=1,inplace=True)

all_pitchers = get_all_bp(stats,1)

#performing more data cleaning. Here, there is an issue of FP/G for some pitches, which must be fixed
all_pitchers = all_pitchers[(all_pitchers['FP/G'] != '[?]') & (all_pitchers['FP/G'] != '-[?]')]

stats_to_int = ['Rk', 'Age', 'GP', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'BB', 
                'SO', 'SB', 'CS', 'RC', 'TB', 'XBH', 'PA', 'year']
pstats_to_int = ['Rk', 'Age', 'W', 'SV', 'QS', 'BS', 'WP', 'HB', 'IBB', '1B', '2B', '3B', 
                 'HR', 'GS', 'L', 'H', 'R', 'ER', 'BB', 'K', 'CG', 'SHO', 'HLD','year']

stats_to_float = ['FP/G', 'FPts','AVG', 'OBP', 'SLG', 'OPS', 'BB/PA', 'K/PA', 'ISO', 'BABIP', 'WOBA']
pstats_to_float = ['FPts', 'FP/G', 'IP', 'K/9', 'BB/9', 'HR/9', 'GB/FB', 'HR/FB', 
                   'BABIPA', 'ERA', 'LOB%', 'FIP', 'W%', 'BR/9', 'BAA', 'WHIP']

for stat in stats_to_int:
    convert_to_int(all_batters,stat)

for stat in stats_to_float:
    convert_to_float(all_batters,stat)
    
for stat in pstats_to_int:
    convert_to_int(all_pitchers,stat)

for stat in pstats_to_float:
    convert_to_float(all_pitchers,stat)
```

Now, we must tidy and clean the data. First, it was necessary to add a column to each dataframe representing the year from which the data came. Then, we need to filter out the batters who did not receive enough at-bats in a season. For the years 2016-2019, any player with fewer than 125 at-bats was removed. However in 2020 there was a shortened season due to the coronavirus pandemic, so the threshold for at-bas was lowered to 50. After this, we modified the age column for both the batters and pitchers. Initially, the age column represented the current age of the player, regardless of the value in the year column. Thus, a quick modification changes the values in the age column to represent the true age of a player during a season. Finally, we combine all of the batter data into one dataframe, all_batters, and all of the pitcher data into another dataframe, all_pitchers and converted each column type to be more representative of the true column type of the data in the column.

```python
all_pitchers = all_pitchers[all_pitchers['IP'] > 0]
all_pitchers['G'] = all_pitchers.apply(lambda row : (row['FPts'] / row['FP/G']) if row['FP/G'] != 0 else -1,axis=1)
#A PLAYER IS A RP IF THEY PLAYED AS RP IN MORE THAN 2/3 OF THEIR GAMES
all_pitchers['Position'] = all_pitchers.apply(lambda row: row['Position'] if row['G'] == -1 else
                                              'SP' if row['G'] <= 3*row['GS'] else 'RP',axis=1)
all_starters = copy.deepcopy(all_pitchers[all_pitchers['Position'] == 'SP'])
all_relievers = copy.deepcopy(all_pitchers[all_pitchers['Position'] == 'RP'])

all_starters=all_starters[(all_starters['year'] == 2020) & (all_starters['IP'] >= 20) | 
                          (all_starters['year'] != 2020) & (all_starters['IP'] >= 50)]
    
all_relievers = all_relievers[(all_relievers['year'] == 2020) & (all_relievers['IP'] >= 10) |
               (all_relievers['year'] != 2020) & (all_relievers['IP'] >= 25)]
```

Lastly, it is important to note that there are two types of ptichers, starters (SP) and relievers (RP). Since starters and relievers have different roles in baseball and as a result earn points in a slightly different manner, they must be analyzed separately. A pitcher was considered a starter if he started at least 1/3 of the games in which he appeared. After separating starters and relievers, it was necessary to enforce a playing time requirement, which is innings pitched for pitchers. For starters, they must have pitched at least a total of 50 innings in the years 2016-2019 and at least 20 innings in 2020. Relievers, on the other hand, pitched at lot less than starters, so their innings pitched threshold is lower (25 innings for the years 2016-2019 and 10 innings for 2020). By enforcing a playing time requirement for both batters and pitchers, we take out data observations for players that played a very small amount in a season, which could significantly alter the results of the analysis.

<h2>Exploratory Data Analysis</h2>

To be able to predict what batters or pitchers will perform well, it would be helpful to understand the distribution of points scored as well as what statistics most heavily influence the number of points scored. In the following histograms, we display the points per game for batters, starting pitchers, and relief pitchers. It is important to make a distinction between starting and relief pitchers because starting pitchers, by the definition of their job, will accrue more overall playing time and thus accumulate more statistics and fantasy points than relief pitchers, who only pitch for a very short part of a game.

```python
convert_to_float(all_batters,'FP/G')
convert_to_float(all_starters,'FP/G')
convert_to_float(all_relievers,'FP/G')

ppg_batters = all_batters['FP/G']
ppg_starters = all_starters['FP/G']
ppg_relievers = all_relievers['FP/G']

#creating the distributions
fig, ax = plt.subplots(1, 3,figsize=(15,8))
ax[0].hist(ppg_batters,bins=50)
ax[0].set_xlabel('Fantasy Points Per Game')
ax[0].set_ylabel('Number of Players')
ax[0].set_title('Fantasy Points Per Game for Batters')

ax[1].hist(ppg_starters,bins=50)
ax[1].set_xlabel('Fantasy Points Per Game')
ax[1].set_title('Fantasy Points Per Game for Starting Pitchers')

ax[2].hist(ppg_relievers,bins=50)
ax[2].set_xlabel('Fantasy Points Per Game')
ax[2].set_title('Fantasy Points Per Game for Relief Pitchers')

plt.show()
```

In the above histograms, we immediately notice that starting pitchers earn by far the most points per game compared to other positions. With a median of 5 fantasy points per game, the median starting pitcher earns about 2.75 more points per game than the median batter and about 3.5 more points per game than the median relief pitchers. However, one must take into consideration that batters, where many play every day, see the field significantly more often than starting pitchers, who typically pitch only once every 5 days. So even though a starting pitcher is much more likely to earn a higher number of points in a given game, batters will likely earn more points over a single fantasy matchup, which typically last a week, sometimes longer. That is not to say batters will always be more valuable than starting pitchers, nor that relief pitchers have zero value compared to the other positions. Rather, it helps provide insight into when a batter might be more valuable than a starting pitcher and vice versa. Additionally, relief pitchers, especially the ones who score the most points per game, can provide a lot of value to a team, as they pitch more often than starting pitchers.

Even though the above distribtuions does not directly relate to the question of predicting a player's future performance, they do provide a basic background into the points generated by each group of players. This, along with an understanding of the playing times of the players, allows fantasy owners to determine how much to generally invest in each position group.

```python
batters_ab = all_batters['AB']
starters_ip = all_starters['IP']
relievers_ip = all_relievers['IP']

#creating the scatter plots
fig, ax = plt.subplots(1, 3,figsize=(15,8))
ax[0].scatter(batters_ab,ppg_batters)
ax[0].set_xlabel('At Bats')
ax[0].set_ylabel('Fantasy Points Per Game')
ax[0].set_title('FP/G v. AB for Batters')

ax[1].scatter(starters_ip,ppg_starters)
ax[1].set_xlabel('Innings Pitched')
ax[1].set_ylabel('Fantasy Points Per Game')
ax[1].set_title('FP/G v. IP for Starting Pitchers')

ax[2].scatter(relievers_ip,ppg_relievers)
ax[2].set_xlabel('Innings Pitched')
ax[2].set_ylabel('Fantasy Points Per Game')
ax[2].set_title('FP/G v. IP for Relief Pitchers')

plt.show()
```

Another important thing to point out is the relationship between playing time and points per game. For batters and starting pitchers, and not so much relief pitchers, there appears to be a moderate positive linear relationship between playing time and points per game. While this appears to be something to potentially explore, the amount of at-bats has little to no extent a cause of a player's points per game being higher or lower. While more at-bats in an individual game would surely increase the opportunities for a player to earn more points, the data points included in the above scatter plots detail a player's performance over the course of a season. Since the data is long-term, it is much more likely and intutive that a batters who performs better, somewhat reflected in their points per game, will earn more at-bats. Thus, at bats cannot be used to predict the points per game earned by batters.

However, innings pitched directly impacts a pitcher's points per game, as the scoring system awards one point to a pitcher for each inning pitched. With a higher innings pitched, a pitcher will accumulate more points. Also, a pitcher who throws more is likely doing so due to a positive pitching outing. Thus, innings pitched can be used to predict the overall performance level of a starting pitcher. However, the graph of points per game v. innings pitched for relief pitchers shows almost no linear relationship between the two variables, so innings pitched cannot be reliably used to predict the performance of a relief pitcher.

Now, we must determine what factors best predict a player's performance, as we cannot use all of the statistics in our model to predict player performance. In the linear regression model, one of the assumptions is that there does not exist perfect collinearity between the variables used in the model. For example, if both hits and batting average have a linear relationship with each other, then the linear regression model cannot include both of hits and batting average. Thus, it is important to explore what might impact a player's performance in fantasy baseball and select the variables that appear to most strongly impact performance.

We first start with the age of a position player (batter). A player's age, in any sport, could impact their performance because as players get older, they gain more experience and knowledge in the game. However, their bodies are expected to be less able to handle the rigors of a full 162-game season as they get older. With that in mind, we plotted the violin plots above. Contrary to expectations, there does not seem to be, with a few exceptions, much of an impact of age on points scored per game. Most of the ages have a mean points per game of around 2 except for players aged 20, 21, 39, or 40, the two youngest and two oldest ages. A likely explanation for this is since few baseball players earn significant playing time at those young ages and the ones that do are typically expected to be among the best players in baseball in their careers, the sample size is too small and biased towards those uber-talented players. Additionally, the distribution for players aged 39 and 40 contains a small number of players such that we cannot properly determine the distribution of points per game for players of those ages

It is also important to point out the distribution of the points per game as seen in the above plots. There appears to be two sets of ages, 22-28 and 29-34 with similar distributions while the rest of the years have their own unqiue distributions. The top left plot above plots the relationship between points per game and batting average.
